```python
#2.1 Изучите распределение домов от наличия вида на набережную, Постройте график
data = df['waterfront'].value_counts()
plt.pie(data.values,labels=data.index,autopct='%1.2f%%')
plt.title('наличие вида на набережную')
plt.legend(['нет','есть']);
```

```python
df1 = df['floors'].value_counts()
plt.bar(df1.index, df1.values);
plt.xlabel('этажность в доме')
plt.ylabel('кол-во домов')
plt.title('распределение кол-во домов по этажам');
```

```python
plt.hist(df['condition'],bins=100);
plt.xlabel('оценка состояния')
plt.ylabel('кол-во домов')
plt.title('распределение домов по оценке состояния');
```

Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.boxplot(y=df['price'],x= df['bedrooms'], whis = 4,showfliers = False)
plt.ylabel('Цена дома*100000')
plt.xlabel('кол-во комнат')
plt.title('Зависимость цены от кол-во комнат');
```

```python
plt.scatter(df['yr_built'],df['price'])
plt.ylabel('Цена дома*100000')
plt.xlabel('год постройки')
plt.title('Зависимость цены от года постройки');
```

```python
df2 = df[['price','view','grade']]
matr = round(df2.corr(), 2)
sns.heatmap(matr, annot = True, linewidths=.5, cmap='coolwarm');
plt.title('Зависимость цены от вида и уровня');
```
